Our task involves classifying the different species in the Iris dataset with the help of Decision Tree Classifier

# SANKALP JAIN

Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('Iris.csv')
df
```

```python
# Dropping Id column as it has no use
df.drop(columns='Id',inplace=True)
df.head()
```

Checking for any missing values in the dataset


```python
df.isna().sum()
```

Separating label from the dataset and converting it to numerical

```python
Y=df['Species']
Y.value_counts()
```

```python
Y=Y.map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
Y
```

```python
X=df.drop(columns='Species')
X.head()
```

Training the Decision Tree Classifier

```python
model=DecisionTreeClassifier(random_state=21)
model.fit(X,Y)
```
